# Lendo e analisando arquivo Domicilio01_RN:


```python
domicilio01_RN = pd.read_excel("data/Domicilio01_RN.XLS")
print(domicilio01_RN.shape)
domicilio01_RN.head()
```

```python
domicilio01_RN = domicilio01_RN.replace({'X': np.nan})
domicilio01_RN.head()
```

Lendo e analisando arquivo Basico_RN:



```python
basico_RN = pd.read_excel("data/Basico_RN.xls")
print(basico_RN.info())
print(basico_RN.shape)
basico_RN.head()
```

```python
new_columns = []
for col in domicilio01_RN.columns:
    if(col[:1] == 'V'):
        name = 'Domicilio01_RN_' + col
        new_columns.append(name)
    else:
        new_columns.append(col)
domicilio01_RN.columns = new_columns
domicilio01_RN = domicilio01_RN.drop('Situacao_setor', axis=1)
print(domicilio01_RN.shape)
domicilio01_RN.head()
```

Juntando os arquivos basico_RN e domicilios01_RN

```python
domicilios_rn = basico_RN.merge(domicilio01_RN, on='Cod_setor')
print(domicilios_rn.info())
print(domicilios_rn.shape)
domicilios_rn.head()
```

Filtrando para municipio de Natal


```python
domicilios_natal = domicilios_rn[domicilios_rn.Nome_do_municipio == 'NATAL'].reset_index(drop=True)
print(domicilios_natal.info())
print(domicilios_natal.shape)
```

```python
domicilios_natal.columns.values
```

```python
print(len(domicilios_natal.Nome_do_bairro.unique()))
domicilios_natal.Nome_do_bairro.unique()
```

```python
#domicilios_natal = domicilios_natal.drop(domicilios_natal.iloc[:, 0:33],axis = 1)
domicilios_natal = domicilios_natal.drop(columns=['Cod_Grandes Regiões', 'Nome_Grande_Regiao', 'Cod_UF', 'Nome_da_UF ', 
                                                  'Cod_meso', 'Nome_da_meso', 'Cod_micro', 'Nome_da_micro', 'Cod_RM', 'Nome_da_RM', 
                                                  'Cod_municipio', 'Nome_do_municipio', 'Cod_distrito', 'Nome_do_distrito', 'Cod_subdistrito', 
                                                  'Nome_do_subdistrito', 'Cod_bairro', 'Situacao_setor', 'Tipo_setor', 'V001', 'V002', 'V003', 'V004', 
                                                  'V005', 'V006', 'V007', 'V008', 'V009', 'V010', 'V011', 'V012']).sort_values('Nome_do_bairro').reset_index(drop=True)
print(domicilios_natal.info())
print(domicilios_natal.shape)
domicilios_natal.head()
domicilios_natal.to_csv("domicilios_natal.csv", index=False, encoding="UTF-8")
```

# Análise: Tipo de domicílios x bairros(Natal) x Número de banheiros x Tipo de esgotamento

O passo inicial das listas de tarefas nesta unidade era que cada aluno precisava escolher pelo menos uma análise ou pergunta específica para projetar um gráfico na segunda etapa. O objetivo principal era iniciar com a comparação de variáveis e, em seguida, explorar como o mesmo gráfico poderia apoiar análises de identificação e sumarização dos dados e variáveis.

A análise das variáveis mencionadas, como a quantidade de domicílios, bairros em Natal, tipo de domicílios, número de banheiros e tipo de esgotamento, pode ser interessante por várias razões:

* Planejamento urbano: ao analisar a quantidade de domicílios por bairros em Natal, é possível identificar padrões de distribuição populacional e entender a demanda habitacional em cada região. Num contexto de administração, esse tipo de análise poderia auxiliar no planejamento urbano, na definição de políticas públicas e na alocação adequada de recursos para infraestrutura e serviços.

* Qualidade de vida: a análise do tipo de domicílios e do número de banheiros pode ajudar a compreender sobre a qualidade de vida da população. Domicílios com maior número de banheiros tendem a proporcionar mais conforto e melhores condições sanitárias. Essa informação pode ser relevante para identificar áreas com deficiências na infraestrutura básica.

* Saneamento básico: o tipo de esgotamento está diretamente relacionado à questão do saneamento básico. Ao analisar essa variável, é possível identificar bairros com maior ou menor acesso a sistemas de esgotamento sanitário adequados. Esse tipo de informação num contexto de definição de políticas públicas é importante para direcionar investimentos e ações visando melhorar o saneamento básico e, consequentemente, a saúde e qualidade de vida da população.

* Desigualdades sociais: a análise das variáveis mencionadas pode revelar desigualdades sociais e disparidades entre diferentes bairros da cidade. Por exemplo, a concentração de domicílios com menor número de banheiros em determinados bairros pode indicar condições precárias de moradia e desigualdades socioeconômicas.
