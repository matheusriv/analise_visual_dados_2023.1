```python
import pandas as pd
import numpy as np
```

# Lendo e analisando arquivo responsavelRenda_natal:

```python
# lendo arquivo basico e tirando colunas
basico_RN = pd.read_excel("../data/Basico_RN.xls")\
              .drop(columns=['Cod_Grandes Regiões', 'Nome_Grande_Regiao', 'Cod_UF', 'Nome_da_UF ',
                            'Cod_meso', 'Nome_da_meso', 'Cod_micro', 'Nome_da_micro', 'Cod_RM', 'Nome_da_RM',
                            'Cod_municipio', 'Cod_distrito', 'Nome_do_distrito', 'Cod_subdistrito',
                            'Nome_do_subdistrito', 'Cod_bairro', 'Situacao_setor', 'Tipo_setor',
                            'V001', 'V002', 'V003', 'V004', 'V005', 'V006', 'V007', 'V008',
                             'V009', 'V010', 'V011', 'V012'])
```

```python
basico_RN.head()
```

```python
# lendo tabela responsavelRenda
responsavelRenda_natal = pd.read_excel("../data/ResponsavelRenda_RN.XLS").replace({'X': np.nan}).drop('Situacao_setor', axis=1)
new_columns = []
# renomeando colunas
for col in responsavelRenda_natal.columns:
    if(col[:1] == 'V'):
        name = 'ResponsavelRenda_RN_' + col
        new_columns.append(name)
    else:
        new_columns.append(col)
responsavelRenda_natal.columns = new_columns
# juntando com a tabela basico_RN para obter nomes dos bairros
responsavelRenda_natal = basico_RN.merge(responsavelRenda_natal, on='Cod_setor')
```

```python
print(responsavelRenda_natal.info())
print(responsavelRenda_natal.shape)
responsavelRenda_natal.head()
```

```python
# filtrando para municipio de natal
responsavelRenda_natal = responsavelRenda_natal[responsavelRenda_natal.Nome_do_municipio == 'NATAL'].reset_index(drop=True)
# retirando coluna nome do municipio
responsavelRenda_natal = responsavelRenda_natal.drop('Nome_do_municipio', axis=1)
responsavelRenda_natal.head()
```

```python
# nomes dos bairros de natal
print(len(responsavelRenda_natal.Nome_do_bairro.unique()))
responsavelRenda_natal.Nome_do_bairro.unique()
```

```python
# colocando em ordem alfabética dos bairros
responsavelRenda_natal = responsavelRenda_natal.sort_values('Nome_do_bairro')
# agrupando por bairro e somando linhas, retirando coluna Codigo do setor
responsavelRenda_natal = responsavelRenda_natal.groupby('Nome_do_bairro', as_index=False).sum().drop(columns=['Cod_setor'])
```

```python
print(responsavelRenda_natal.shape)
responsavelRenda_natal.head()
```

# Lendo outras tabelas úteis:

```python
#lendo tabela responsavel01
responsavel01_natal = pd.read_excel("../data/Responsavel01_RN.xls").replace({'X': np.nan}).drop('Situacao_setor', axis=1)
new_columns = []
# renomeando colunas
for col in responsavel01_natal.columns:
    if(col[:1] == 'V'):
        name = 'Responsavel01_RN_' + col
        new_columns.append(name)
    else:
        new_columns.append(col)
responsavel01_natal.columns = new_columns
# juntando com tabela basico_RN para obter nomes dos bairros e filtrando para Natal
responsavel01_natal = basico_RN.merge(responsavel01_natal, on='Cod_setor')
responsavel01_natal = responsavel01_natal[responsavel01_natal.Nome_do_municipio == 'NATAL'].reset_index(drop=True).drop('Nome_do_municipio', axis=1)
# agrupando por bairro e somando linhas, retirando coluna Codigo do setor
responsavel01_natal = responsavel01_natal.sort_values('Nome_do_bairro')
responsavel01_natal = responsavel01_natal.groupby('Nome_do_bairro', as_index=False).sum().drop(columns=['Cod_setor'])

#lendo tabela responsavel02
responsavel02_natal = pd.read_excel("../data/Responsavel02_RN.xls").replace({'X': np.nan}).drop('Situacao_setor', axis=1)
new_columns = []
# renomeando colunas
for col in responsavel02_natal.columns:
    if(col[:1] == 'V'):
        name = 'Responsavel02_RN_' + col
        new_columns.append(name)
    else:
        new_columns.append(col)
responsavel02_natal.columns = new_columns
# juntando com tabela basico_RN para obter nomes dos bairros e filtrando para Natal
responsavel02_natal = basico_RN.merge(responsavel02_natal, on='Cod_setor')
responsavel02_natal = responsavel02_natal[responsavel02_natal.Nome_do_municipio == 'NATAL'].reset_index(drop=True).drop('Nome_do_municipio', axis=1)
# agrupando por bairro e somando linhas, retirando coluna Codigo do setor
responsavel02_natal = responsavel02_natal.sort_values('Nome_do_bairro')
responsavel02_natal = responsavel02_natal.groupby('Nome_do_bairro', as_index=False).sum().drop(columns=['Cod_setor'])
```

```python
#lendo tabela domicilio02
domicilio02_natal = pd.read_excel("../data/Domicilio02_RN.xls").replace({'X': np.nan}).drop('Situacao_setor', axis=1)
new_columns = []
# renomeando colunas
for col in domicilio02_natal.columns:
    if(col[:1] == 'V'):
        name = 'Domicilio02_RN_' + col
        new_columns.append(name)
    else:
        new_columns.append(col)
domicilio02_natal.columns = new_columns
# juntando com tabela basico_RN para obter nomes dos bairros e filtrando para Natal
domicilio02_natal = basico_RN.merge(domicilio02_natal, on='Cod_setor')
domicilio02_natal = domicilio02_natal[domicilio02_natal.Nome_do_municipio == 'NATAL'].reset_index(drop=True).drop('Nome_do_municipio', axis=1)
# agrupando por bairro e somando linhas, retirando coluna Codigo do setor
domicilio02_natal = domicilio02_natal.sort_values('Nome_do_bairro')
domicilio02_natal = domicilio02_natal.groupby('Nome_do_bairro', as_index=False).sum().drop(columns=['Cod_setor'])
```
